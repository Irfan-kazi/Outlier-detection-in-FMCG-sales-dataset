```python
def detect_outliers_zscore(df,threshold=3):
    z_scores=stats.zscore(df.select_dtypes(include=[np.number]))
    abs_z_scores=np.abs(z_scores)
    outliers=(abs_z_scores>threshold).any(axis=1)
    return df[outliers]
    
```
